```python
import numpy as np
import pandas as pd

df_DEG = pd.read_csv(r'string_data/string_DEG_S2.tsv',delimiter='\t')

arr_df_DEG_gene = []
for i in df_DEG['#node1'].unique():
    arr_df_DEG_gene.append(i)
print(len(arr_df_DEG_gene))
print(arr_df_DEG_gene[:10])
```

```python
# Create a function to solve the repeated elements in the list：
def kill_pardon(list_nokill):
    f_list = []
    for item in list_nokill:
        if item not in f_list:
            f_list.append(item)
    return f_list
```

```python
for i in df_DEG.node2.unique():
    arr_df_DEG_gene.append(i)
arr_df_DEG_gene = kill_pardon(arr_df_DEG_gene)
print(len(arr_df_DEG_gene))
print(arr_df_DEG_gene[:10])
```

```python
# Create an empty dataframe：
def create_df_ini(arr_matrix_col):
    form_header = ['']
    for item in arr_matrix_col:
        form_header.append(item)
    df = pd.DataFrame(columns = form_header)
    df[''] = arr_matrix_col
    return df

# Create the adjacent matrix：
def create_matrix(arr_matrix_col,df_in_form,df_ini):
    cnt = 0
    arr = []
    for item in df_in_form.columns:
        arr.append(item)
    ele0 = arr[0]
    ele1 = arr[1]
    if ele1 == 'lnc_DGE':
        ele1 = arr[2]
    
    print(ele0,end=" ")
    print(ele1)
    
    x = 0
    for elem_m in df_in_form[ele0]:
        elem_l = df_in_form[ele1][cnt]
        cnt = cnt + 1
        
        # flag: mark the position of the gene in the matrix(the column and row information)
        if elem_l in arr_matrix_col and elem_m in arr_matrix_col:
            flag_m = arr_matrix_col.index(elem_m)
            flag_l = arr_matrix_col.index(elem_l)
            
            x = x + 1
            df_ini.at[flag_m,elem_l] = 1
            df_ini.at[flag_l,elem_m] = 1
    
    print(x)
    return df_ini

# Create a function to make diagonal elements in the matrix equals to 0：
def set_zero(arr_matrix_col,df):
    #set the diagonal equals to 0，that is gene can not reach itself：
    for i in range(0,len(arr_matrix_col)):
        temp = arr_matrix_col[i]
        for j in range(0,len(df[temp])):
            if i == j:
                df.at[j,temp] = 0
    #return the dataframe：
    return df
```

```python
# Create the adjacent matrix：
dataframe2 = create_df_ini(arr_matrix_columns)
# input df1 to df3 to fill dataframe2, this step only fill 1 in the matrix:
dataframe2 = create_matrix(arr_matrix_columns,df1,dataframe2)
cnt = 0
arr_no_one = []
for elem in arr_matrix_columns:
    list_cnt = []
    list_cnt = dataframe2[elem]
    for item in list_cnt:
        if item == 1:
            cnt = cnt+1
print(cnt/2)
dataframe2 = create_matrix(arr_matrix_columns,df2,dataframe2)
cnt = 0
arr_no_one = []
for elem in arr_matrix_columns:
    list_cnt = []
    list_cnt = dataframe2[elem]
    for item in list_cnt:
        if item == 1:
            cnt = cnt+1
print(cnt/2)
dataframe2 = create_matrix(arr_matrix_columns,df3,dataframe2)
cnt = 0
arr_no_one = []
for elem in arr_matrix_columns:
    list_cnt = []
    list_cnt = dataframe2[elem]
    for item in list_cnt:
        if item == 1:
            cnt = cnt+1
print(cnt/2)
dataframe2 = create_matrix(arr_matrix_columns,df_DEG,dataframe2)
cnt = 0
arr_no_one = []
for elem in arr_matrix_columns:
    list_cnt = []
    list_cnt = dataframe2[elem]
    for item in list_cnt:
        if item == 1:
            cnt = cnt+1
print(cnt/2)

# make the diagonal 0：
dataframe2 = set_zero(arr_matrix_columns,dataframe2)
#dataframe2.info()
```
