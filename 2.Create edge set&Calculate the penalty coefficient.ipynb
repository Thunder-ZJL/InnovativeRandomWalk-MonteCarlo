```python
import numpy as np
import pandas as pd

df_DAM = pd.read_csv('string_data/D_adjacency_matrix.csv')
arr_col = []
for i in df_DAM.columns:
    arr_col.append(i)
arr_col.remove(arr_col[0])
df_DAM
```

```python
dict_degree = {}
for item in arr_col:
    count = 0
    point_list = df_DAM[item]
    for elem in point_list:
        if elem == 1:
            count = count + 1
    dict_degree[item] = count

dict_degree
```

```python
cnt = 0
for key,value in dict_degree.items():
    cnt = cnt + value
cnt
```

# Build the edge setï¼š

```python
# a new dataframe to get the edge setï¼š
df_bianji = pd.DataFrame(columns = ['Source','Target'])
l_S = list_T = []
for i in range(len(arr_col)):
    pA = arr_col[i]
    pB_list = []
    pB_temp = []
    pB_temp = df_DAM.iloc[i,2:]
    for j in range(i,len(pB_temp)):
        if pB_temp[j] == 1:
            pB_list.append(arr_col[j+1])
    if len(pB_list) != 0:
        for item in pB_list:
            list_T.append(item)
df_bianji['Target'] = list_T
```

```python
l_S = []
for i in range(len(arr_col)):
    pA = arr_col[i]
    pB_list = []
    pB_temp = []
    pB_temp = df_DAM.iloc[i,2:]
    for j in range(i,len(pB_temp)):
        if pB_temp[j] == 1:
            pB_list.append(arr_col[j])
    if len(pB_list) != 0:
        for k in range(len(pB_list)):
            l_S.append(pA)
            #print(l_S)

df_bianji['Source'] = l_S
```

# Calculate the penalty coefficient of every vertex based on the degreeï¼š

```python
df_edge = pd.read_csv('string_data/EDGE_of_D_adjMatrix.csv')
df_edge
```

```python
dict_WeightPenalty = {}
for item in arr_col:
    dict_WeightPenalty[item] = 0
```

```python
for i in range(len(df_edge)):
    point_s = df_edge['Source'][i]
    point_t = df_edge['Target'][i]
    value1 = dict_degree[point_t]
    value2 = dict_degree[point_s]
    dict_WeightPenalty[point_s] = dict_WeightPenalty[point_s] + value1
    dict_WeightPenalty[point_t] = dict_WeightPenalty[point_t] + value2

dict_WeightPenalty
```

```python
import math

cf_list = {}
lg_list = []
for key,value in dict_WeightPenalty.items():
    if value != 0:
        cf_list[key] = 0.5 * math.log(value) + 1
cf_list
```
