```python
def run_matrix_by_RandomWalk_withCF(df,path,arr_matrix_col,steps,times,cf_list):
    # Rwalk is the routine list in random walk:
    #Rwalk = []
    #Rroad is the result in each round walk:
    Rroad = {}
    # initiate the list of the basic weight：
    lamda_list = []
    for i in range(1,steps+1):
        lamda = steps - i + 1
        lamda_list.append(lamda)
    fg = 0
    # first pick the start (need to choose every vertex as the start)
    for t in range(0,times):
        start_point_index = 0
        for i_start in range(0,len(arr_matrix_col)):
            start_point_index = i_start
            # every round should initiate the Rwalk list:
            Rwalk = []
            # set a WALK: in the network walk in 'steps' steps:
            for i in range(0,steps):
                # put the start into the Rwalk list:
                # NOTICE!: 100 round corresponding to count>=100, and the start need to refresh later
                Rwalk.append(arr_matrix_col[start_point_index])
                
                # cnt: count 1 in a line;
                # temp_list: storage the adjacent vertices of the current vertex
                # temp_road: storage the indexs of the adjacent vertices corresponding to 1
                cnt = 0
                temp_road = []
                temp_list = []
        
                for j in arr_matrix_col:
                    temp_list.append(df[j][start_point_index])
                
                for k in range(0,len(temp_list)):
                    if temp_list[k] == 1:
                        cnt = cnt + 1
                        temp_road.append(k)
                
                if cnt != 0:
                    fg = 1
                    # generate a random number to choose current routine, that is select the number of vertices as the next path choice:
                    random_number = np.random.randint(0,cnt)
                    
                    # the index needs to be updated with the line as the target
                    start_point_index = temp_road[random_number]
            
            if fg == 1:
                # the endpoint is added to Rwalk list at the end of the last loop
                Rwalk.append(arr_matrix_col[start_point_index])
                # integrating weights in sequence:
                Rwalk_dict = {}
                # Add penalty coefficient when integrating weights, final weight=basic_weight * penalty_coefficient
                for ind in range(0,steps):
                    Rwalk_dict[Rwalk[ind]] = lamda_list[ind] * cf_list[Rwalk[ind]]
                # put the result of walk: Rwalk in the dictionary: Rroad 
                Rroad[arr_matrix_col[i_start]] = Rwalk_dict
                fg = 0
        
        # integrating weights：
        for key,value in Rroad.items():
            # get the dictionary in path[key]:
            dict_k = path[key]
            #print(value)
            for k,v in value.items():
                #print(value)
                if k not in dict_k:
                    dict_k[k] = v
                else:
                    value1 = dict_k[k]
                    value2 = v + value1
                    dict_k[k] = value2
            # put the integrated dict_k back to path：
            path[key] = dict_k
        
        # make Rroad empty：
        Rroad = {}
        print("\r["+str(t+1)+" round is over!!]",end = '\n')
    
    # return the dictionary：
    return path    
```

```python
# function for selection：
def select_point_bylam(pathx,lamda):
    # only retain the vertices >= lamda
    pathy = {}
    for key,value in pathx.items():
        point_list = []
        for k,v in value.items():
            if v >= lamda:
                point_list.append(k)
        pathy[key] = point_list
    return pathy

def jiao_and_bing(arr_matrix_col,path_over_half_times,dict_jiaoji,dict_bingji):
    #### Create the union and the intersection：
    # Use union/intersection to select:
    for key,value in dict_jiaoji.items():
        str_tmp1 = key.split('@')[0]
        str_tmp2 = key.split('@')[1]
        list_A = path_over_half_times[str_tmp1]
        list_B = path_over_half_times[str_tmp2]
        jiao = []
        for item in list_A:
            if item in list_B:
                jiao.append(item)
        # put A∩B into key：
        for item in jiao:
            value.append(item)
    print("jiao is OK!")
    
    for key,value in dict_bingji.items():
        str_tmp1 = key.split('@')[0]
        str_tmp2 = key.split('@')[1]
        list_A = path_over_half_times[str_tmp1]
        list_B = path_over_half_times[str_tmp2]
        bing = []
        for item in list_A:
            bing.append(item)
        for item in list_B:
            bing.append(item)
        bing = kill_pardon(bing)
        # put A∪B into key：
        for item in bing:
            value.append(item)
    print("bing is OK!")
    
    #### Test for the value of the union/intersection:
    # union/intersection >=0.5：
    point_list_1 = []
    # union/intersection >=0.45：
    point_list_2 = []
    # union/intersection >=0.4:
    point_list_3 = []
    # union/intersection >=0.35:
    point_list_4 = []
    # union/intersection >=0.3:
    point_list_5 = []
    # union/intersection >=0.25:
    point_list_6 = []
    # union/intersection >=0.2:
    point_list_7 = []
    # union/intersection >=0.15:
    point_list_8 = []
    # union/intersection >=0.1:
    point_list_9 = []
    
    # door as the threshold：union/intersection selection:
    door = [0.9,0.8,0.7,0.6,0.5,0.4,0.3,0.2,0.1]
    point_list_1 = control_jiao_and_bing(dict_jiaoji,dict_bingji,door[0])
    print(len(point_list_1))
    dict_1 = combine_bingji(point_list_1,arr_matrix_col)
    print(dict_1)
    print("------------------------------------------------------------")
    point_list_2 = control_jiao_and_bing(dict_jiaoji,dict_bingji,door[1])
    print(len(point_list_2))
    dict_2 = combine_bingji(point_list_2,arr_matrix_col)
    print(dict_2)
    print("------------------------------------------------------------")
    point_list_3 = control_jiao_and_bing(dict_jiaoji,dict_bingji,door[2])
    print(len(point_list_3))
    dict_3 = combine_bingji(point_list_3,arr_matrix_col)
    print(dict_3)
    print("------------------------------------------------------------")
    point_list_4 = control_jiao_and_bing(dict_jiaoji,dict_bingji,door[3])
    print(len(point_list_4))
    dict_4 = combine_bingji(point_list_4,arr_matrix_col)
    print(dict_4)
    print("------------------------------------------------------------")
    point_list_5 = control_jiao_and_bing(dict_jiaoji,dict_bingji,door[4])
    print(len(point_list_5))
    dict_5 = combine_bingji(point_list_5,arr_matrix_col)
    print(dict_5)
    print("------------------------------------------------------------")
    point_list_6 = control_jiao_and_bing(dict_jiaoji,dict_bingji,door[5])
    print(len(point_list_6))
    dict_6 = combine_bingji(point_list_6,arr_matrix_col)
    print(dict_6)
    print("------------------------------------------------------------")
    point_list_7 = control_jiao_and_bing(dict_jiaoji,dict_bingji,door[6])
    print(len(point_list_7))
    dict_7 = combine_bingji(point_list_7,arr_matrix_col)
    print(dict_7)
    print("------------------------------------------------------------")
    point_list_8 = control_jiao_and_bing(dict_jiaoji,dict_bingji,door[7])
    print(len(point_list_8))
    dict_8 = combine_bingji(point_list_8,arr_matrix_col)
    print(dict_8)
    print("------------------------------------------------------------")
    point_list_9 = control_jiao_and_bing(dict_jiaoji,dict_bingji,door[8])
    print(len(point_list_9))
    dict_9 = combine_bingji(point_list_9,arr_matrix_col)
    print(dict_9)
#     print("------------------------------------------------------------")
#     point_list_10 = control_jiao_and_bing(dict_jiaoji,dict_bingji,door[9])
#     print(len(point_list_10))
#     dict_10 = combine_bingji(point_list_10,arr_matrix_col)
#     print(dict_10)

# function for combine the intersection：
def combine_bingji(point_list,arr_matrix_col):
    # a dictionary to record the clusters:
    dict_point={}
    for item in arr_matrix_col:
        dict_point[item]='null'
    # flag: record the No. of current cluster：
    flag = 1
    # after traversing the corresponding elements each time, the corresponding value is modified to the number after clustering：
    for i in point_list:
        # before adding: check all the preceding clusters:
        cluster = []
        for j in i:
            if dict_point[j] != 'null':
                cluster.append(dict_point[j])
        # check the vertices: whether the clusters should be combined together
        if len(cluster) >= 1:
            # make elements no repeated and the value !=1 means apper cluster information >=2: 
            cluster = kill_pardon(cluster)
            if len(cluster) > 1:
                # flagx：record the min number of the cluster:
                flagx = 1000000
                for elem in cluster:
                    if elem < flagx:
                        flagx = elem
                # traverse: modifies the large cluster number to the small cluster number：
                for key,value in dict_point.items():
                    if value in cluster:
                        dict_point[key] = flagx
                # pay attention to the previous flag:
                flag = flag -1
            # =1 means do not need to modify the number：
            else:
                idx = cluster[0]
                for j in i:
                    dict_point[j] = idx
        # if cluster number does not contain current number:
        else:
            #print("here goes else")
            for j in i:
                dict_point[j] = flag
                # flag+1 after every loop：
            flag = flag + 1
            
    # intergrate the clusters：
    flagxx = []
    list_f = []
    for key,value in dict_point.items():
        if value != 'null':
            flagxx.append(value)
    flagxx = kill_pardon(flagxx)
    for item in flagxx:
        list_i = []
        for key,value in dict_point.items():
            if value == item:
                list_i.append(key)
        #print(list_i)
        list_f.append(list_i) 
    # use the secondary list to make the cluster dictionary:
    dict_a = {}
    for i in range(0,len(list_f)):
        name = "C"+str(i+1);
        dict_a[name] = list_f[i]
    return dict_a

# function to control union/intersection：
def control_jiao_and_bing(dict_jiaoji,dict_bingji,door):
    # create an empty point_list to get the selected point_list：
    point_list = []
    for key,value in dict_jiaoji.items():
        # get A∩B and A∪B：
        point_list1 = value
        point_list2 = dict_bingji[key]
        # A∩B/A∪B
        len1 = len(point_list1)
        len2 = len(point_list2)
        # percent = union/intersection：
        if len2 != 0:
            percent = len1*1.0/len2
        else:
            percent = 0
            
        # use door to do the selection：
        if percent >= door:
            # if >= door: put the intersection into the point_list to make a secondary list:
            point_list.append(point_list2)
        
    return point_list
```
